### 1. Question: Merge and Sort Two Lists

**Description:** Given two lists of integers, merge them into one list and sort the result in ascending order. Do not use Python's built-in sort() function or other sorting functions.




```python
'''
# Example Input:
list1 = [3, 1, 4]
list2 = [2, 5, 0]

# Example Output:
[0, 1, 2, 3, 4, 5]'''
```

```python
l1 = input("Enter the first list: ")
l2 = input("Enter the second list: ")
l1 = [int(x) for x in l1.split()]
l2 = [int(x) for x in l2.split()]
l1 += l2
for i in range(len(l1)):
  for j in range(len(l1)):
    if l1[i] < l1[j]:
      l1[i], l1[j] = l1[j], l1[i]
print(l1)
```
